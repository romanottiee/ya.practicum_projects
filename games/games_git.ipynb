В моем распоряжении данные из открытых источников о продажах игр, оценках пользователей и экспертов, жанрах и платформах.
Моя задача — выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании для интернет-магазина "Стримчик".

Цель исследования:
1. Определить наиболее перспективные платформы и актуальный период для построения прогноза;
2. Составить портрет пользователя каждого региона;
3. Сформулировать и проверить гипотезы.

Ход исследования:

Данные о недвижимости я получу из файла `/datasets/games.csv`. О качестве данных ничего не известно. Поэтому перед тем, как приступать к целям исследования, необходимо выполнить обзор данных.

Я проверю данные на ошибки и оценю их влияние на исследование. На этапе обработки я поищу возможность исправить все ошибки данных, которые не приведут к искажению конечного результата. Далее создам необходимые столбцы и приступлю к проведению исследовательского анализа.

Таким образом, мое исследование пройдет в шесть этапов:
- Обзор данных;
- Предобработка данных и дополнительные расчеты;
- Исследовательский анализ данных;
- Составление портрета пользователя в зависимости от региона;
- Формулировка и проверка гипотез;

### Предобработка данных

#### Приведение заголовков к нижнему регистру

````python
data.columns = data.columns.str.lower()
````

````python
data.head()
````

#### Удаление пропусков

````python
data.isna().sum() #посчитаем количество пропусков в данных по столбцам
````

##### Устранение пропусков в столбцах 'name' и 'genre'

В указанных столбцах есть по два пропуска. Проверим, может это одни и те же строки?

````python
data[data['name'].isna()]
````

Да, так и есть. Избавимся от них, поскольку нельзя исследовать неизвестно чего.

````python
data = data.loc[data['name'].notna()]
````

##### Устранение пропусков в столбце 'year_of_release'

В столбце 'year_of_release' количество пропусков составляет 269. Есть два пути: отыскать недостающую информацию и заполнить вручную либо избавиться от этих строк. Выбираем второй вариант, поскольку количество строк с пропущенными данными едва ли повлияет на общую картину.

````python
data = data.loc[data['year_of_release'].notna()]
````

##### Устранение пропусков в столбцах 'critic_score', 'user_score' и 'rating'

К сожалению, поступить аналогично предыдущему с указанными столбцами мы не можем, иначе в данных будут неприемлемые потери. Но что же предпринять?
- Пропуски в столбцах 'critic_score' и 'user_score' можно заполнить отрицательными значениями, а затем их фильтровать. Или же оставить пропуски. Выбираем второй вариант;
- Пропуски в столбце 'rating' заменим на 'No rating'.

````python
data['rating'] = data['rating'].fillna('No rating')
````

Следует обратить внимание на значение 'tbd' в столбце 'user_score'. Посмотрим, что это.

````python
data[data['user_score']=='tbd']
````

Беглый взгляд на датафрейм позволяет предположить, что такую пометку имеют некоммерческие игры (нулевые или околонулевые продажи) или игры на непопулярных платформах. Скорее всего, сама пометка означает 'to be defined' (англ. "уточняется", "появится позже"). Возможно, количества оценок на агрегаторе, с которого снята статистика, недостаточно для расчета оценки. Заменим это обозначение на пропуск для того, чтобы можно было изменить тип данных данного столбца.

````python
data['user_score'] = data['user_score'].replace('tbd',np.nan)
````

````python
data.isna().sum() #проверка пропусков после выполненных преобразований
````

Все корректно.

Можно предположить, что имевшиеся пропуски появились из-за того, что в базах-агрегаторах данных попросту не было той или иной информации (например, неправильно создана карточка, или недосточно данных для расчета оценок, или игра слишком старая, поэтому у нее нет возрастного рейтинга - ESRB основана в 1994 году, или иная причина).

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```
H_0: Средний пользовательский рейтинг (user_score) платформы Xbox One = Средний пользовательский рейтинг (user_score) платформы PC
H_a: Средний пользовательский рейтинг (user_score) платформы Xbox One ≠ Средний пользовательский рейтинг (user_score) платформы PC
alpha = 0.05
```

````python
results = st.ttest_ind(
    selected_period_data.loc[((data['platform'] =='XOne')&(data['user_score'].notna())), 'user_score'],
    selected_period_data.loc[((data['platform'] =='PC')&(data['user_score'].notna())), 'user_score'],
    equal_var = True
)
    
alpha = 0.05

# вывод значения p-value на экран 

print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
````

Формулировка нулевой и альтернативной гипотезы следует из самой задачи. Здесь мы сравниваем две независимые генеральные совокупности, поэтому метод проверки - ttest_ind. Т.к. нулевая гипотеза предполагает равенство среднего двух совокупностей, то в качестве аргумента equal_var следует поставить True.

Вывод: результат теста говорит о том, что отвергнуть нулевую гипотезу не пполучилось.

#### Средние пользовательские рейтинги жанров Action и Sports разные

```
H_0: Средний пользовательский рейтинг (user_score) жанра Action = Средний пользовательский рейтинг (user_score) жанра Sports
H_a: Средний пользовательский рейтинг (user_score) жанра Action ≠ Средний пользовательский рейтинг (user_score) жанра Sports
alpha = 0.05
```

````python
results = st.ttest_ind(
    selected_period_data.loc[((data['genre'] =='Action')&(data['user_score'].notna())), 'user_score'],
    selected_period_data.loc[((data['genre'] =='Sports')&(data['user_score'].notna())), 'user_score'],
    equal_var = True
)
    
alpha = 0.05

# вывод значения p-value на экран 

print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
````

Несмотря на иную формулировку, задача идентична предыдущей, поэтому нулевая гипотеза также предполагает равенство. При этом мы также сравниваем две независимые генеральные совокупности, поэтому тот же метод ttest_ind.

Вывод: на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы..
